```python
# Импортируем библиотеку NumPy с общепринятым сокращением
import numpy as np

# Пример 1: Создание массива
list_a = [1, 3, 5, 7, 9]
a = np.array(list_a)
print("Вывод:", a)

# Пример 2: Создание массива с использованием функции zeros
b = np.zeros(4) 
print("Вывод:", b)

# Пример 3: Создание массива с использованием функции ones
c = np.ones(5)  
print("Вывод:", c)

# Пример 4: Создание массива с использованием функции arange
d = np.arange(0, 8, 2)  # 
print("Вывод:", d)

# Пример 5: Демонстрация атрибутов
print("Форма массива a:", a.shape)  
print("Тип данных массива a:", a.dtype)  

# Пример 6: Индексирование и срезы
element = a[2]  # Получение отдельного элемента (индексация)
print("Вывод:", element)

subarray = a[1:4]  # Извлечение подмассива с помощью среза
print("Вывод:", subarray)

# Пример 7: Векторные операции и базовая статистика
e = np.array([1, 3, 7, 8, 9])
sum_arrays = a + e  # Поэлементное сложение
print("Вывод:", sum_arrays)

mul_arrays = a * e  # Поэлементное умножение
print("Вывод:", mul_arrays)

mean_a = np.mean(a) # Вычисление средней величины
print("Вывод:", mean_a)

sum_a = np.sum(a) # Вычисление суммы всех элементов массива
print("Вывод:", sum_a)

std_a = np.std(a) # Вычисление стандартного отклонения
print("Вывод:", std_a)
```
